````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

**Figure**: The top-level container for all plot elements, which can include multiple Axes

The three arguments in `fig.add_subplot(nrows, ncols, index)` represent the following:

> nrows: The number of rows in the subplot grid.

> ncols: The number of columns in the subplot grid.

> index: The position of the current subplot in the grid, using a 1-based index.

So, when calling fig.add_subplot(1, 1, 1), you are creating a subplot grid with only one row and one column, and you are placing the current subplot in the first (and only) position of that grid.

````python
# Create a figure
fig = plt.figure()

# Add an axes to the figure
ax = fig.add_subplot(1, 1, 1)

# Plot data on the axes x and y axis
ax.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])

# Set labels for the axes
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

# Set a title for the plot
ax.set_title("A simple line plot")

# Display the plot
plt.grid()
plt.show()
````

````python
# Create a figure
fig = plt.figure(figsize=(8, 8))

# Add the first subplot (top-left) to the figure
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
ax1.grid()
ax1.set_title('Line Plot')

# Add the second subplot (top-right) to the figure
ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
ax2.set_title('Scatter Plot')

# Add the third subplot (bottom-left) to the figure
ax3 = fig.add_subplot(2, 2, 3)
ax3.bar(['A', 'B', 'C', 'D', 'E'], [3, 5, 7, 6, 2])
ax3.set_title('Bar Plot')

# Add the fourth subplot (bottom-right) to the figure
ax4 = fig.add_subplot(2, 2, 4)
ax4.hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4], bins=[1, 2, 3, 4, 5])
ax4.set_title('Histogram')

# Adjust the layout of subplots to avoid overlap
plt.tight_layout()

# Display the figure with all subplots
plt.show()
````

In this example, we create a single figure and then use the `add_subplot()` method to create a 2x2 grid of subplots (4 subplots in total). The three arguments in `fig.add_subplot(nrows, ncols, index)` are used to specify the grid configuration and the position of the current subplot:

---
    nrows: The number of rows in the grid (2 in this case).
    ncols: The number of columns in the grid (2 in this case).
    index: The position of the current subplot in the grid (from 1 to 4 in this example).
---

For each subplot, we create a different type of plot (line plot, scatter plot, bar plot, and histogram) and add a title to the subplot using the `set_title()` method.

Finally, we use the plt.tight_layout() function to adjust the layout of the subplots to avoid any overlaps, and then display the figure containing all subplots using `plt.show()`.

**Axes**: The actual plot, containing the labels, lines, and other visual elements.
Axes are the actual plots within a Figure. They contain the labels, lines, and other visual elements. You can create Axes using the `fig.add_subplot()` method or by using the `plt.subplots()` function.

Example 1: Using `fig.add_subplot()` to create a single Axes:

````python
# Create a figure
fig = plt.figure()

# Add an axes to the figure, create a grid of 1 row, 1 col
ax = fig.add_subplot(1, 1, 1)

# Plot data on the axes
ax.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])

# Display the plot
plt.show()
````

````python
# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2)

# Add data to the subplots
axes[0, 0].plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
axes[0, 1].scatter([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
axes[1, 0].bar([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
axes[1, 1].hist([0, 1, 2, 2, 3, 3, 3, 4, 4, 4, 4])

# Display the plots
plt.show()
````

When using the pyplot interface, Matplotlib automatically creates a Figure object for you if you don't create one yourself. When you call functions like `plt.plot()`, `plt.scatter()`, or `plt.bar()` directly, Matplotlib uses the current (or default) figure and axes to create the plot.

However, if you need more control over the figure properties or want to create multiple figures or subplots, using `plt.figure()` or `plt.subplots()` is recommended. This allows you to manage the properties of the Figure and Axes objects explicitly, which is useful when creating more complex visualizations.

````python
x = np.random.rand(50)
y = np.random.rand(50)

fig = plt.figure(figsize=(6, 6))  # Create a new figure with a custom size
plt.scatter(x, y, marker='*', s=100, c='red', edgecolor='black')
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

````python
# same results, previous example, matplotlib automatically manages Figure and Axes

fig, ax = plt.subplots(figsize=(6, 6))  # Create a new figure and axes with a custom size
ax.scatter(x, y, marker='*', s=100, c='red', edgecolor='black')
ax.set_title('Scatter Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
````

````python
# managing two figures directly within plt 

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create the first figure and plot
plt.figure(1, figsize=(6, 4))
plt.plot(x, y1)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Create the second figure and plot
plt.figure(2, figsize=(6, 4))
plt.plot(x, y2)
plt.title('Cosine Wave')
plt.xlabel('x')
plt.ylabel('cos(x)')

# Show both figures
plt.show()
````

````python
x_series = pd.Series(np.arange(1,6))
y_series = pd.Series([y ** 2 for y in range(1, 6)])

plt.plot(x_series, y_series)
plt.xlabel("x series")
plt.ylabel("y series")
plt.title("Simple Line Plot")
plt.show()
````

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas

# Data for the plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a new figure and a canvas to render the figure
fig = Figure()
canvas = FigureCanvas(fig)

# Create an axis object and add it to the figure
ax = fig.add_subplot(111)

# Plot the data using the axis object
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Simple Line Plot')

# Display the plot
fig.canvas.draw()
fig.show()
```

As you can see, the matplotlib.pyplot version abstracts away the object creation and management, like creating a Figure object, adding an Axes object, and rendering the figure on a canvas. This makes it easier to use and results in less code compared to the object-oriented approach.

In summary, matplotlib.pyplot hides the object creation and management under the hood, providing a simpler API for creating and manipulating plots. It automatically takes care of creating and managing Figure and Axes objects, allowing you to focus on defining your data and plotting it with minimal code.
