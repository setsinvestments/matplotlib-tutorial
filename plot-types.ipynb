```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

In this example, we create a 3x2 grid of subplots using `plt.subplots()` and plot a different type of plot in each subplot. We have included examples of 

1. line plots
2. scatter plots
3. bar plots
4. histograms
5. pie charts

The sample data for each plot is generated using NumPy and Python lists. 
We use the `set_title()` method to add a title to each subplot, 
and the `plt.tight_layout()` function to adjust the layout to avoid overlaps.

---

Note that there are additional plot types available in Matplotlib, 
such as box plots, violin plots, contour plots, and 3D plots. 
These plot types are more specialized and may require additional parameters or data formatting.

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D', 'E']
values = [3, 5, 7, 6, 2]
data = np.random.randn(1000)
pie_labels = ['Apple', 'Banana', 'Orange', 'Grape']
pie_sizes = [40, 30, 20, 10]

# Create a figure
fig, axes = plt.subplots(3, 2, figsize=(10, 15))

# Line plot
axes[0, 0].plot(x, y)
axes[0, 0].set_title('Line Plot')

# Scatter plot
axes[0, 1].scatter(x, y)
axes[0, 1].set_title('Scatter Plot')

# Bar plot
axes[1, 0].bar(categories, values)
axes[1, 0].set_title('Bar Plot')

# Histogram
axes[1, 1].hist(data, bins=30)
axes[1, 1].set_title('Histogram')

# Pie chart
axes[2, 0].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90)
axes[2, 0].axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
axes[2, 0].set_title('Pie Chart')

# Remove unused subplot
fig.delaxes(axes[2, 1])

# Adjust the layout of subplots
plt.tight_layout()

# Display the figure with all subplots
plt.show()
```

**Line Plot** `plt.plot()`:

Real-life example: Visualizing the trend of stock prices over time.

A line plot can be used to visualize how the stock price of a company changes over a period of time. The x-axis could represent the time (days, weeks, months, or years), and the y-axis would represent the stock price.

```python
# Sample data: time and stock prices
time = [0, 1, 2, 3, 4, 5]
stock_prices = [100, 110, 120, 90, 95, 105]

plt.plot(time, stock_prices)
plt.xlabel('Time (days)')
plt.ylabel('Stock Price ($)')
plt.title('Stock Price Trend')
plt.show()
```

**Scatter Plot** `plt.scatter()`:

Real-life example: Visualizing the relationship between age and income.

A scatter plot can be used to explore the correlation between two variables, such as age and income. The x-axis could represent age, and the y-axis would represent income.

```python
# Sample data: age and income
age = [25, 30, 35, 40, 45, 50, 55, 60]
income = [30000, 35000, 40000, 45000, 50000, 55000, 60000, 65000]

plt.scatter(age, income)
plt.xlabel('Age (years)')
plt.ylabel('Income ($)')
plt.title('Age vs. Income')
plt.show()
```

**Bar Plot** `plt.bar()`:

Real-life example: Comparing the population of different cities.

A bar plot can be used to compare the population of different cities, providing a clear visualization of the differences between them. The x-axis would represent the city names, and the y-axis would represent the population.

```python
# Sample data: cities and population
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
population = [8405837, 3884307, 2718782, 2195914, 1513367]

plt.bar(cities, population)
plt.xlabel('City')
plt.ylabel('Population')
plt.title('Population of Major US Cities')
plt.show()
```

**Histogram** `plt.hist()`:

Real-life example: Analyzing the distribution of test scores in a class.

A histogram can be used to visualize the distribution of test scores in a class, providing insights into the overall performance of the students. The x-axis would represent the score ranges, and the y-axis would represent the number of students in each score range.

```python
# Sample data: test scores
test_scores = [75, 80, 85, 90, 80, 85, 90, 95, 100, 80, 85, 90, 95, 100, 85, 90, 95, 100, 85, 90, 95, 100]

# bin is the width of the bar
plt.hist(test_scores, bins=[70, 80, 90, 100], edgecolor='black')

plt.xlabel('Test Score Range')
plt.ylabel('Number of Students')
plt.title('Distribution of Test Scores')
plt.show()
```
